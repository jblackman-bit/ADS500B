In this exercise, we practice writing Python functions and loops. 
    4.1: Define a Python function is_in_high_school(age) that takes the age of a person and determines if the person is in high school or not. Assume that for a person to be in high school, their age should be between 14 and 18, inclusive. (4 points) 
Explanation
    4.2: Iterate over the list [25,18,9,13,34,15,22,17,12,37,15] and call your function is_in_high_school(age) on each age value. (4 points) 
Explanation
    4.3: Calculate the percentage of people in your age list not going to high school. (4 points)
Explanation


```python
### Question 4.1
min_high_school_age = 14
max_high_school_age = 18

def is_in_high_school(age):
    return min_high_school_age <= age <= max_high_school_age

### Question 4.2 and 4.3
age_list = [25, 18, 9, 13, 34, 15, 22, 17, 12, 37, 15]
not_high_school_count = 0

for i in age_list:
    
    if is_in_high_school(i):
        print("You are in high school because you are " + str(i))
    else:
        not_high_school_count += 1
        print("You are not in high school because you are " + str(i))

not_high_school_percentage = round((not_high_school_count / len(age_list)) * 100, 2)
print(str(not_high_school_percentage) + "% of people are not in high school.")
```
